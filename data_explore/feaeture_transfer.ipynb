```python
# 定义全局函数便于直接调用处理
from datetime import datetime

import pandas as pd
import numpy as np
from datetime import date
import datetime as dt
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import boxcox
datapath = 'D:/outsourcing/data/'
featurepath = 'D:/outsourcing/feature/'
resultpath = 'D:/outsourcing/result/'
tmppath = 'D:/outsourcing/tmp/'
```

```python
combine_train = pd.read_csv(tmppath+'/sub/combine_feature_train.csv')
combine_test = pd.read_csv(tmppath+'/sub/combine_feature_test.csv')
```

```python
combine_values_train = combine_train.values[:,44:-1]
combine_values_test = combine_test.values[:,44:]#去除所有原来的数据与lable标签
```

```python
combine_features = combine_train.columns.tolist()[44:-1]
print(combine_features)
```

```python
feature_make_train = pd.read_csv(tmppath+'/sub/feature_make_train.csv')
feature_make_test = pd.read_csv(tmppath+'/sub/feature_make_test.csv')
```

```python
feature_make_train_values = feature_make_train.values
feature_make_test_values = feature_make_test.values
```

```python
feature_make_features = feature_make_train.columns.tolist()
print(feature_make_features)
```

```python
train = pd.DataFrame(np.concatenate([feature_make_train_values, combine_values_train], axis=1))

df_train_all = pd.DataFrame(train)
df_train_all.columns = feature_make_features + combine_features
df_train_all.to_csv('sub_train_all.csv', header=True, index=False)
```

```python
test = pd.DataFrame(np.concatenate([feature_make_test_values, combine_values_test], axis=1))
feature_make_features = feature_make_test.columns.tolist()
df_train_all = pd.DataFrame(test)
df_train_all.columns = feature_make_features + combine_features
df_train_all.to_csv('sub_test_all.csv', header=True, index=False)
```
